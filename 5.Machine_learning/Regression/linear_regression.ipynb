```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
df=pd.read_csv('salary.data.csv')
df.head(6)
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['WC'])
df.head(6)
```

```python
plt.scatter(df['EDU'],df['SAL'],color='blue')
plt.xlabel("EDU")
plt.ylabel("SAL")
plt.title("Salary_Data")
plt.show()
```

```python
df['SAL']
```

```python
model=LinearRegression()
Education=np.array(df["EDU"]).reshape(-1,1)
print(Education)
```

```python
salary=np.array(df['SAL'])
print(salary)
```

```python
model.fit(Education,salary)
model_predict=model.predict(Education)
```

```python
table = np.column_stack((salary,model_predict))

print(tabulate(table[:5], headers =("Actual Data","Predicted Data"), tablefmt = 'fancy_grid'))
```

```python
plt.scatter(df['EDU'],df['SAL'])
plt.plot(df['EDU'],model_predict,color='r',label="Salary_of_BFL")
plt.xlabel('EDU')
plt.ylabel('SAL')
plt.title('Salary on Education')
plt.legend()
plt.show()
```

```python
R_squared=np.round(r2_score(salary,model_predict),4)
print("R_squared:",R_squared)
```
